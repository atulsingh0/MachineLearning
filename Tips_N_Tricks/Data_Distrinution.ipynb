```python
# import
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
normed_data = pd.DataFrame(np.random.normal(size=1000))

# plot density, median and mean 
normed_data.plot(kind='density')
#plt.plot(normed_data.mean())

plt.vlines(normed_data.mean(),     # Plot black line at mean
           ymin=0, 
           ymax=0.4,
           linewidth=7.0)

plt.vlines(normed_data.median(),     # Plot black line at mean
           ymin=0.0, 
           ymax=0.4,
           linewidth=5.0,
          color="red")
```
